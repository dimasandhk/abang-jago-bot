## ***SOAL 1 Modul 1.1***

```python
# 1. Diketahui tinggi badan siswa kelas A adalah
# 100,5 ; 99,7 ; 101 ; 101,3 ; 99,9 ; 98,9 ; 165 ; 102,2 ; 98,7 ; 102,9 ; 100
# Menggunakan Mean, Median, dan Modus, gambarkan pusat distribusi data berikut.
# Manakah yang paling bagus?

data = [100.5, 99.7 ,101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]
print(f'Data tinggi badan siswa kelas A: {data}')

# MEAN
jumlahTinggi = sum(data) # Jumlah semua data
panjangData = len(data) # Dapatkan panjang data
meanTinggi = jumlahTinggi / panjangData
print('\nMEAN:')
print(f'Mean = membagi jumlah tinggi seluruh siswa ({jumlahTinggi:.2f})', end=' ')
print(f'dengan banyak siswa = {panjangData} siswa')
print(f'Mean dari tinggi badan siswa kelas A adalah {meanTinggi:.2f}')
```

```python
# MEDIAN
# Mengurutkan data rendah -> tinggi
sortedData = sorted(data)
print('MEDIAN:')
print("Untuk mencari median perlu mengurutkan data terlebih dahulu (ascending): ")
print(sortedData)

tengah = len(data) // 2
medianTinggi = 0

dataTengahGenap = sortedData[tengah] + sortedData[tengah + 1]
medianTinggi = sortedData[tengah] if panjangData % 2 else dataTengahGenap / 2
print(f'Median dari data tersebut adalah {medianTinggi}')
```

```python
# MODUS
# Modus yang saya ambil dari tinggi siswa tanpa koma
# karena sedekat apapun tingginya ke angka selanjutnya (99.9) tingginya tetap dianggap 99 bukan 100
import math
from collections import Counter

dataBulat = [math.floor(num) for num in data]
dataJumlahTinggi = dict(Counter(dataBulat))

jumlahTerbanyak = max(dataJumlahTinggi.values())
modusTinggi = [key for key, jumlah in dataJumlahTinggi.items() if jumlah == jumlahTerbanyak]

print(f'Data yang telah dibulatkan: {dataJumlahTinggi}')
print(f'Modus dari tinggi badan siswa kelas A adalah {modusTinggi}')
```

```python
import matplotlib.pyplot as plt
# Buat Histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='yellow')

# Buat Garis penanda Mean
plt.axvline(x=meanTinggi, color='green', linestyle='dashed', linewidth=1, label=f'Mean = {meanTinggi:.2f}')
# Penanda Median
plt.axvline(x=medianTinggi, color='red', linewidth=3, label=f'Median = {medianTinggi}')
# Penanda Modus
for modus in modusTinggi:
  plt.axvline(x=modus, color='purple', linestyle='dashed', linewidth=1, label=f'Modus = {modus}')

# Buat title dan label
plt.xlabel('Tinggi')
plt.ylabel('Frekuensi')
plt.title('Data Tinggi Badan Siswa Kelas A')
plt.figtext(0.5, -0.05, f'Data Tinggi: {data}', fontsize=12, color='black', ha='center')

# Tambahkan legenda dan tunjukkan
plt.legend()
plt.show()
```

## ***Soal 2 Modul 1.1***

```python
# Berikut adalah nilai UAS pada mata pelajaran matematika di kelas A:
# 95, 35, 50, 65, 70, 85, 25, 100, 80, 90, 45, 40, 35, 20, 75, 50, 40, 60, 80, 75.
# Menggunakan Mean, Median, dan Modus, coba gambarkan pusat distribusi data berikut. Manakah yang paling bagus?
dataNilai = [95, 35, 50, 65, 70, 85, 25, 100, 80, 90, 45, 40, 35, 20, 75, 50, 40, 60, 80, 75]
print(f'Data Nilai UAS MAT Kelas A: {dataNilai}\n')

# MEAN
jumlahNilai = sum(dataNilai)
panjangDataNilai = len(dataNilai)
meanNilai = jumlahNilai / panjangDataNilai

print('MEAN:')
print(f'Mean = membagi jumlah nilai seluruh siswa ({jumlahNilai:.2f})', end=' ')
print(f'dengan banyak siswa = {panjangDataNilai} siswa')
print(f'Mean dari Nilai UAS Kelas A adalah {meanNilai:.2f}')
```

```python
# MEDIAN
# Mengurutkan data rendah -> tinggi
sortedDataNilai = sorted(dataNilai)
print('MEDIAN:')
print("Untuk mencari median perlu mengurutkan data terlebih dahulu (ascending): ")
print(sortedDataNilai)

tengahNilai = len(dataNilai) // 2
medianNilai = 0

nilaiTengahGenap = sortedDataNilai[tengahNilai] + sortedDataNilai[tengahNilai + 1]
medianNilai = sortedDataNilai[tengahNilai] if panjangDataNilai % 2 else nilaiTengahGenap / 2
print(f'Median dari data tersebut adalah {medianNilai}')
```

```python
# MODUS
from collections import Counter

dataJumlahNilai = dict(Counter(dataNilai))
jumlahNilaiTerbanyak = max(dataJumlahNilai.values())
modusNilai = [key for key, jumlah in dataJumlahNilai.items() if jumlah == jumlahNilaiTerbanyak]

print(f'Data nilai: {dataJumlahNilai}')
print(f'Modus dari nilai UAS MAT kelas A adalah {modusNilai}')
```

```python
# Buat Histogram
plt.hist(dataNilai, bins=30, edgecolor='black', alpha=0.5, color='yellow')

# Buat Garis penanda Mean
plt.axvline(x=meanNilai, color='green', linestyle='dashed', linewidth=1, label=f'Mean = {meanNilai:.2f}')
# Penanda Median
plt.axvline(x=medianNilai, color='red', linewidth=3, label=f'Median = {medianNilai}')
# Penanda Modus
for modus in modusNilai:
  plt.axvline(x=modus, color='purple', linestyle='dashed', linewidth=1, label=f'Modus = {modus}')

# Buat title dan label
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.title('Data Nilai UAS MAT Kelas A')
plt.figtext(0.5, -0.05, f'Data Nilai: {dataNilai}', fontsize=12, color='black', ha='center')

# Tambahkan legenda dan tunjukkan
plt.legend()
plt.show()
```
